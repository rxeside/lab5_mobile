# 14. Предварительный анализ данных. Графическое представление

```python
# Import the libraries
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import seaborn as sns
```

Загрузим данные о продолжительности извержений гейзеров.

```python
#создание DataFrame (df) с помощью чтения данных из файла
df = pd.read_excel('C:/Users/admin/Downloads/01/01/geyser.xlsx')
```

Построим **гистограмму** с относительными частотами.

```python
plt.hist(df['duration'], color = 'blue', edgecolor = 'black', weights=np.ones_like(df)/len(df), bins = int(210/8))
plt.xlabel('Продолжительность извержения в секундах')
# сохранение гистограммы
plt.savefig('geyser', dpi = 200, transparent=True, bbox_inches = 'tight')
```

Построим **гистограмму** с частотами

```python
plt.hist(df['duration'], color = 'blue', edgecolor = 'black', bins = int(210/9));
plt.xlabel('Ширина = 9');
plt.savefig('geyser', dpi = 200, transparent=True, bbox_inches = 'tight')
```

Построим **ядерную оценку плотности**

```python
sns.kdeplot(df['duration'], kernel='epa', bw=10, legend=False);
plt.title("Ядро Епанечникова и ширина = 10");
plt.savefig('kernel', dpi = 200, transparent=True, bbox_inches = 'tight')
```

```python
sns.kdeplot(df['duration'], kernel='gau', bw=10, legend=False);
plt.title("  и ширина = 10");
plt.savefig('kernel', dpi = 200, transparent=True, bbox_inches = 'tight')
```

```python
sns.kdeplot(df['duration'], kernel='triw', bw=10, legend=False);
plt.title("Триквадратное ядро и ширина = 10")
plt.savefig('kernel', dpi = 200, transparent=True, bbox_inches = 'tight')
```

```python
sns.kdeplot(df['duration'], kernel='epa', bw=24, legend=False, color = 'blue');
sns.kdeplot(df['duration'], kernel='gau', bw=24, legend=False, color = 'red');
sns.kdeplot(df['duration'], kernel='triw', bw=24, legend=False, color = 'green');
plt.legend(('epa', 'gau', 'triw'))
plt.title("Разные ядра и ширина = 24");
plt.savefig('kernel', dpi = 200, transparent=True, bbox_inches = 'tight')
```

**Ядра с границами**

```python
df_time = pd.read_excel('C:/Users/admin/Downloads/01/01/inter failure times.xlsx')
```

```python
sns.kdeplot(df_time['time'], kernel='gau', bw='silverman', legend=False, color = 'blue');
plt.title("Особенности симметричного ядра");
plt.savefig('inter failure times', dpi = 200, transparent=True, bbox_inches = 'tight')
```

```python
#pip install KDEpy
import KDEpy
from KDEpy import FFTKDE
```

```python
data = df_time['time'].values
kde = FFTKDE(bw='silverman', kernel='triweight')
x, y = kde.fit(data)(135)

low_bound = 0
data = np.concatenate((data, 2 * low_bound - data))
x, y = FFTKDE(bw=kde.bw, kernel='triweight').fit(data)(270)
y[x<=low_bound] = 0
y = y * 2

plt.xlim(0, 7000)
plt.plot(x, y, color = 'blue')
plt.savefig('inter failure times', dpi = 200, transparent=True, bbox_inches = 'tight')
```
